```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('titanic.csv')
df
```

```python
df1 = df[['Pclass','Sex','Age','Fare','Survived']]
df1
```

```python
df1.shape
```

```python
df1.isnull().sum()
```

```python
df1['Age'] = df1['Age'].fillna(df1['Age'].mean())
```

```python
df1.isnull().sum()
```

```python
df1.shape
```

```python
df2 = df1.dropna()
```

```python
df2.shape
```

```python
df2
```

```python
dummies = pd.get_dummies(df2.Sex)
```

```python
dummies
```

```python
df3 = pd.concat([df2,dummies],axis=1)
```

```python
df3
```

```python
df3.drop(['Sex'], axis =1,inplace=True)
```

```python
df3
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test = train_test_split(df3.drop(['Survived'],axis =1),df3['Survived'],test_size=0.2)
```

```python
X_train
```

```python
from sklearn.naive_bayes import GaussianNB
```

```python
model =GaussianNB()
```

```python
model.fit(X_train,y_train)
```

```python
model.score(X_test,y_test)
```

```python
model.predict_proba(X_test[:10])
```
